```python
from glob import glob
import os
import matplotlib.pyplot as plt
from flab.rosetta.rosetta_output.resfile import ResFile
```

```python
resfiles=glob('/home/labs/fleishman/dinal/kemp_elim/noPSSM_filterS_afterRefCat3/*UM*/designable*') #all resfiles ddg from -6 to +6
```

```python
#order the resfiles by ddg 
from itertools import groupby

resfiles.sort(key=lambda x:x.split('/')[-1])

res = [list(i) for j, i in groupby(resfiles, 
                  lambda a: a.split('/')[-1])] 
```

```python
#plot all resfiles by number of possible mutations
import itertools

#stuff for plotting
fig, ax = plt.subplots(5,4)
subs=list(itertools.product(range(5),range(4)))
p=0
#order resfiles so figures will go by order 
res= sorted(res, key=lambda x:float(x[1].split('designable_aa_resfile.')[-1]))

#plot resfiles
for ddg in res:
    ddgV=ddg[0].split('designable_aa_resfile.')[-1]
    ddg_D={num: 0 for num in range(1, 21)} #dictionary that holds mutation # frewuency
    for i in ddg:
        s=ResFile(i)
        for line in s:
            ddg_D[(len(line[1]))]+=1
    #plot
    ax[subs[p]].bar(list(ddg_D.keys()),list(ddg_D.values()))
    ax[subs[p]].set_title('ddg ' +str(ddgV), x=0.5, y=0.9)
    for i, v in enumerate(ddg_D.values()):
        ax[subs[p]].text(i+0.8 , v , str(v), color='blue', size=12, rotation=45, va='bottom')
    p+=1

fig.set_size_inches(25, 35)
fig.savefig('filterscan_allowedMutN.png', dpi=300, transparent=True)
plt.show()
    
    
```

```python
#print pymol commands to color positions with more than 8 mutations 

for ddg in res:
    ddgV=ddg[0].split('designable_aa_resfile.')[-1]
    if ddgV=='0.0': #working only on ddg 0.0
        ddg_D={num: 0 for num in range(1, 21)}
        for i in ddg:
            s=ResFile(i)
            for line in s:
                if len(line[1])>7: #change "7" to any number you want
                    print ('color atomic, {num} AND resi {frust}'.format(num= i.split('/')[0], frust=line[0][:-1]))
                    print ('show sticks, {num} AND resi {frust}'.format(num= i.split('/')[0], frust=line[0][:-1]))
```
